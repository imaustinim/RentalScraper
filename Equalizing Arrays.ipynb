```python
import json
```

```python
def minOperations(arr, threshold, d):
    new_array = []
    print(arr)
    for i in arr:
        zeros = True
        temp_list = []
        x = 0
        while zeros is True:
            e = int(i/d**x)
            if e is 0:
                zeros = False
                break
            temp_list.append(e)
            x +=1
        new_array.append(temp_list)
    print(new_array)

    dic = {}
    for i in new_array:
        for e in i:
            if e not in dic:
                dic[e] = 1
            elif e in dic:
                dic[e] += 1
    
    print(json.dumps(dic,indent=2))
    flist = []
    for d in dic:
        if dic[d] == threshold:
            flist.append(d)

    fdic = {}
    print(flist)
    for f in flist:
        for i in new_array:
            if f in i:
                if f not in fdic:
                    fdic[f] = i.index(f)
                elif f in fdic:
                    fdic[f] += i.index(f)

    print(json.dumps(fdic, indent=2))
    
    x = fdic[flist[0]]
    for d in fdic:
        if fdic[d] < x:
            x = fdic[d]
            print(d, x)
    return x
            
                    
def Test():
    test_array = [64, 33, 31, 25, 60, 58, 399, 20, 119, 202, 23003, 483, 3049]
    minOperations(test_array, 2, 2)
    
    
if __name__ == "__main__":
    Test()
```
